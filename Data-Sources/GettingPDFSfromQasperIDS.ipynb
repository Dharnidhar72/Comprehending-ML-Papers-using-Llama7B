```python
import csv
from arxiv import query
import requests
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive

# Specify the path to your CSV file with arXiv IDs
csv_file_path = '/content/drive/MyDrive/QasperID/output (1).csv'  # Replace with your file path

# Specify the name of the Google Drive folder to save the PDFs
folder_name = "ArXiv_Qasper_PDFs"  # Replace with your desired folder name

# Initialize Google Drive
gauth = GoogleAuth()
gauth.LocalWebserverAuth()  # Follow the authorization instructions in the console
drive = GoogleDrive(gauth)

# Create the target folder on Google Drive
folder = drive.CreateFile({'title': folder_name, 'mimeType': 'application/vnd.google-apps.folder'})
folder.Upload()
folder_id = folder['id']

# Read the CSV file and extract arXiv IDs
arxiv_ids = []
with open(csv_file_path, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # Replace 'arxiv_id_column' with the actual column name in your CSV file
        arxiv_id = row['arxiv_id_column']
        arxiv_ids.append(arxiv_id)

# Fetch the PDFs for the arXiv IDs and save them to Google Drive
for arxiv_id in arxiv_ids:
    papers = query(id_list=[arxiv_id])

    if papers:
        paper = papers[0]  # Assuming there's only one paper for each ID
        pdf_url = paper.pdf_url

        if pdf_url:
            # Download the PDF
            response = requests.get(pdf_url)

            if response.status_code == 200:
                # Save the PDF to Google Drive
                pdf_file = drive.CreateFile({'title': f'{arxiv_id}.pdf', 'parents': [{'id': folder_id}]})
                pdf_file.content = response.content
                pdf_file.Upload()
                print(f"Saved {arxiv_id}.pdf to Google Drive")
            else:
                print(f"Failed to download {arxiv_id} PDF")
        else:
            print(f"No PDF available for {arxiv_id}")
    else:
        print(f"No paper found for {arxiv_id}")
```

```python
import csv
import requests

# Specify the path to your CSV file with arXiv IDs
csv_file_path = '/content/drive/MyDrive/QasperID/output (1).csv'  # Replace with your file path

# Specify the folder where you want to save the PDFs
output_folder = '/content/drive/MyDrive/Arxiv_Qasper_dataset'  # Replace with your folder path

# Read the CSV file and extract arXiv IDs
arxiv_ids = []
with open(csv_file_path, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # Replace 'arxiv_id_column' with the actual column name in your CSV file
        arxiv_id = row['id']
        arxiv_ids.append(arxiv_id)

# Fetch and save the PDFs for the arXiv IDs
for arxiv_id in arxiv_ids:
    pdf_url = f"https://arxiv.org/pdf/{arxiv_id}.pdf"

    response = requests.get(pdf_url)

    if response.status_code == 200:
        # Save the PDF to the specified folder
        with open(output_folder + '/' + f'{arxiv_id}.pdf', 'wb') as pdf_file:
            pdf_file.write(response.content)
        print(f"Saved {arxiv_id}.pdf to {output_folder}")
    else:
        print(f"Failed to download {arxiv_id} PDF")


```

```python
import os
import matplotlib.pyplot as plt

# Assuming the file contains one file name per line
with open('your_file_with_filenames.txt', 'r') as f:
    file_names = f.read().splitlines()

# Assuming files are in a directory, replace with your directory path
directory_path = '/path_to_your_files_directory/'

# Extract sizes for each file
file_sizes = [os.path.getsize(os.path.join(directory_path, file_name)) / 1024 for file_name in file_names]

# Since there are many files, let's consider showing only a subset for visualization
# Here we'll take every 50th file name for x-axis labels for better visibility
display_file_names = [file_names[i] if i % 50 == 0 else '' for i in range(len(file_names))]

# Create a histogram
plt.figure(figsize=(15,7))
plt.bar(display_file_names, file_sizes, color='teal')

# Rotate x-labels for better readability
plt.xticks(rotation=45, ha='right')
plt.ylabel('File Size (KB)')
plt.title('PDF File Sizes from arXiv')

# Show the plot
plt.tight_layout()
plt.show()
```
